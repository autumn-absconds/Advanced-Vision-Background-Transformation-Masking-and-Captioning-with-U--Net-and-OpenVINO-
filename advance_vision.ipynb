## Image Loading and Preprocessing
This loads the image and do th pre-processing by converting the BRG to RGB format as 
OpenCV reads images in BGR format but the OpenVINO IR model expects images in RGB.

Resize them to 512 x 512, and transpose the dimensions to the format the OpenVINO IR model expects.

We add the mean values to the image tensor and scale the input with the standard deviation. It is called the input data normalization before propagating it through the network. The mean and standard deviation values can be found in the dataloader file in the U^2-Net repository and multiplied by 255 to support images with pixel values from 0-255.

```python
def preprocess_image(image_uri):
    IMAGE_URI = image_uri
    input_mean = np.array([123.675, 116.28, 103.53]).reshape(1, 3, 1, 1)
    input_scale = np.array([58.395, 57.12, 57.375]).reshape(1, 3, 1, 1)

    # Load and preprocess the image
    image = cv2.cvtColor(
        src=load_image(IMAGE_URI),
        code=cv2.COLOR_BGR2RGB,
    )

    resized_image = cv2.resize(src=image, dsize=(512, 512))
    # Convert the image shape to a shape and a data type expected by the network
    # for OpenVINO IR model: (1, 3, 512, 512).
    input_image = np.expand_dims(np.transpose(resized_image, (2, 0, 1)), 0)

    input_image = (input_image - input_mean) / input_scale
    
    return input_image,image
```

## Perform Inference on Image
For the segmentation of Image Load the OpenVINO IR model to OpenVINO Runtime and do inference.

```python
def perform_inference_on_image(input_image):
    # Load the network to OpenVINO Runtime.
    compiled_model_ir = core.compile_model(model=model_ir, device_name=device.value)
    # Get the names of input and output layers.
    input_layer_ir = compiled_model_ir.input(0)
    output_layer_ir = compiled_model_ir.output(0)

    # Do inference on the input image.
    start_time = time.perf_counter()
    result = compiled_model_ir([input_image])[output_layer_ir]
    end_time = time.perf_counter()

    inference_time = end_time - start_time
    fps = 1 / inference_time

    print(
        f"Inference finished. Inference time: {inference_time:.3f} seconds, "
        f"FPS: {fps:.2f}."
    )

    return result
```

## Process Inference Result/Remove Background 
Here image is showed only in parts which were white during the segmentation result.


```python
OUTPUT_DIR="output"
os.makedirs(name=OUTPUT_DIR, exist_ok=True)
```

```python
def process_inference_result(image, result, save_directory):
    # Resize the network result to the image shape and round the values
    # to 0 (background) and 1 (foreground).
    # The network result has (1,1,512,512) shape. The `np.squeeze` function converts this to (512, 512).
    resized_result = np.rint(
        cv2.resize(src=np.squeeze(result), dsize=(image.shape[1], image.shape[0]))
    ).astype(np.uint8)

    # Create a copy of the image and set all background values to 255 (white).
    bg_removed_result = image.copy()
    bg_removed_result[resized_result == 0] = 255

    # Generate a random name for the image
    image_name = str(uuid.uuid4()) + '.jpg'
    
    # Construct the full file path
    save_path = os.path.join(save_directory, image_name)

    # Save the processed image by converting the RGB back to BGR as to not mess with colors.
    cv2.imwrite(filename=str(save_path), img=cv2.cvtColor(bg_removed_result, cv2.COLOR_RGB2BGR))

    # Print the image path
    print(f"Image saved at: {save_path}")

    # Create a link to download the image.
    image_link = FileLink(save_path)
    image_link.html_link_str = "<a href='%s' download>%s</a>"
    display(
        HTML(
            "Download the image by clicking on this link: "
            f"{image_link._repr_html_()}"
        )
    )
    
     #for image captioning 
    outputPath = save_path
    
    return outputPath, bg_removed_result,resized_result
```

## <a id="Background_Blending">Blend option with Background</a>
Here play with the value of alpha , the thing is that the transparency of background and foreground image goes in opposite direction . If background will be more visible the means that foreground will become more transparent and vice versa.

```python
def replace_background_and_display(bg_removed_result, background_file, output_dir="output"):
    os.makedirs(name=output_dir, exist_ok=True)

    # Load the background image
    background_image = cv2.cvtColor(src=load_image(background_file), code=cv2.COLOR_BGR2RGB)
    show_background_image=cv2.resize(src=bg_removed_result, dsize=(background_image.shape[1], background_image.shape[0]))
    # Resize the background image to match the dimensions of the result
    background_image = cv2.resize(src=background_image, dsize=(bg_removed_result.shape[1], bg_removed_result.shape[0]))
   
    # Ensure both images have the same dimensions
    bg_removed_result = cv2.resize(src=bg_removed_result, dsize=(background_image.shape[1], background_image.shape[0]))

    # Blend the images using a weighted sum
    alpha = 0.8  # Adjust the alpha value based on your preference
    blended_image = cv2.addWeighted(bg_removed_result, alpha, background_image, 1 - alpha, 0)

    # Save the generated image
    image_name = str(uuid.uuid4()) + '.jpg'
    save_path = os.path.join(output_dir, image_name)
    cv2.imwrite(filename=save_path, img=cv2.cvtColor(blended_image, cv2.COLOR_RGB2BGR))

    # Display the original image and the image with the new background side by side
    fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 7))
    ax[0].imshow(bg_removed_result)
    ax[1].imshow(background_image)
    ax[2].imshow(blended_image)
    for a in ax:
        a.axis("off")
    plt.show()

    # Create a link to download the image
    image_link = FileLink(save_path)
    image_link.html_link_str = "<a href='%s' download>%s</a>"
    display(
        HTML(
            f"The generated image <code>{image_name}</code> is saved in "
            f"the directory <code>{output_dir}</code>. You can also "
            "download the image by clicking on this link: "
            f"{image_link._repr_html_()}"
        )
    )

    outputPath = save_path
    return outputPath

# Example usage:
BACKGROUND_FILE = "https://img.freepik.com/free-photo/beautiful-nature-landscape-with-river-vegetation_23-2150705820.jpg?t=st=1699856785~exp=1699860385~hmac=5b5b3f828b701dc07ed1eb19f92e87f0e3a72886e5c4a39470b31d59110d3389&w=360"
outputPath = replace_background_and_display(bg_removed_result, BACKGROUND_FILE)
print(outputPath)
```
